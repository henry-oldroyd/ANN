```python
relu = Leaky_Relu()
for _ in range(10):
    num = uniform(-20, 20)
    print(f"relu({[num]})  =  {relu.standard([num])}")
    print(f"relu'({[num]})  =  {relu.derivative([num])}")
```

```python
for _ in range(10):
    length = randint(1, 10)
    vector = [uniform(-20, 20) for _ in range(length)]
    print(f"relu({vector})  =  {relu.standard(vector)}")
    print(f"relu'({vector})  =  {relu.derivative(vector)}")
```

```python
mse = MSE()
print(mse.standard(numpy.array([1, 2, 3]), numpy.array([2, 2, 2])))
print(mse.derivative(numpy.array([1, 2, 3]), numpy.array([2, 2, 2])))
```

```python
mse = MSE()
for _ in range(10):
    length = randint(1, 5)
    Y = numpy.array([uniform(-20, 20) for _ in range(length)])
    true_Y = numpy.array([uniform(-20, 20) for _ in range(length)])
    print(f"mse({Y}, {true_Y})  =  {mse.standard(Y, true_Y)}")
    print(f"mse'({Y}, {true_Y})  =  {mse.derivative(Y, true_Y)}")
```
